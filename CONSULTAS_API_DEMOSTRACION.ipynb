## Ejemplo api colisiones por fecha (01/01/2014 = 01%2F01%2F2014)

```python
import requests
import pandas as pd

url = "http://vps-2671696-x.dattaweb.com:8000/collision/collisions_for_day/01%2F01%2F2014/"
response = requests.get(url)
aux = response.json()["data"]
columnas = response.json()["data"][0].keys()

auxiliar = pd.DataFrame()
auxiliar["id_collision"]                    = [value['id_collision'] for value in aux]
auxiliar['crash_date']                      = [value['crash_date'] for value in aux]
auxiliar['borough']                         = [value['borough'] for value in aux]
auxiliar['latitude']                        = [value['latitude'] for value in aux]
auxiliar['longitude']                       = [value['longitude'] for value in aux]
auxiliar['on_street_name']                  = [value['on_street_name'] for value in aux]
auxiliar['number_of_persons_injured']       = [value['number_of_persons_injured'] for value in aux]
auxiliar['number_of_persons_killed']        = [value['number_of_persons_killed'] for value in aux]

auxiliar
```
