## Choques por tiempo

#### agrego la columna con los dias de la semana, otra para el mes y otra para el año

```python
# Agrego columna dia de la semana
df_vehicle_vehicle_crash_clean['CRASH_DATE'] = pd.to_datetime(df_vehicle_vehicle_crash_clean['CRASH_DATE'])
df_vehicle_vehicle_crash_clean['CRASH_DAY'] = df_vehicle_vehicle_crash_clean['CRASH_DATE'].dt.day_name()
# Agrego columna del mes
df_vehicle_vehicle_crash_clean['CRASH_MONTH'] = df_vehicle_vehicle_crash_clean['CRASH_DATE'].dt.to_period('m').dt.month
# Agrego columna del año
df_vehicle_vehicle_crash_clean['CRASH_YEAR'] = df_vehicle_vehicle_crash_clean['CRASH_DATE'].dt.to_period('Y')

df_vehicle_vehicle_crash_clean.head()
```

```python
df_vehicle_vehicle_crash_clean.head()
```

#### Choques por minuto (historico)

```python
# Creo funciones para renombrar de QTY a PERC
def qty_perc():
  return df_vehicle_vehicle_crash_clean.rename(columns={'QTY':'PERC'}, inplace=True)

def perc_qty():
  return df_vehicle_vehicle_crash_clean.rename(columns={'PERC':'QTY'}, inplace=True)
```

```python
# Renombro la columna UNIQUE_ID a QTY para saber la cantidad
df_vehicle_vehicle_crash_clean.rename(columns={'UNIQUE_ID':'QTY'}, inplace=True)
# Choques por minuto (cantidad)
KPI_veh_crash_minute = df_vehicle_vehicle_crash_clean.groupby('CRASH_TIME')['QTY'].count()
KPI_veh_crash_minute
```

```python
# Choques por minuto (porcentaje)
qty_perc()
KPI_veh_crash_minute_perc = df_vehicle_vehicle_crash_clean.groupby('CRASH_TIME')['PERC'].count()/df_vehicle_vehicle_crash_clean['PERC'].count()*100
KPI_veh_crash_minute_perc
```
